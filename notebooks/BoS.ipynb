```python
from games.bos import BoS
from agents.regretmatching import RegretMatching
from agents.fictitiousplay import FictitiousPlay
from agents.random_agent import RandomAgent
```

```python
g = BoS()
```

```python
g.reset()
```

```python
for agent in g.agents:
    print(f"{agent} reward matrix: {g._R[agent]}")
```

```python
random_agents = dict(map(lambda agent: (agent, RandomAgent(game=g, agent=agent)), g.agents))
```

```python
for i in range(3):
    actions = dict(map(lambda agent: (agent, random_agents[agent].action()), g.agents))
    g.step(actions)
    print(f"Actions: {list(map(lambda agent: g._moves[actions[agent]], g.agents))}")
    for agent in g.agents:
        print(f"{agent} reward: {g.rewards[agent]}")
```

```python
rm = dict(map(lambda agent: (agent, RegretMatching(game=g, agent=agent)), g.agents))
```

```python
from collections import defaultdict
joint_policy = defaultdict(int)
for i in range(10000):
    actions = dict(map(lambda agent: (agent, rm[agent].action()), g.agents))
    key = tuple([int(actions[agent]) for agent in actions.keys()])
    joint_policy[key] += 1
    g.step(actions)
```
